```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')

# Load Kaggle dataset
data = pd.read_csv("C:\\Users\\Shree\\Downloads\\data.csv")
```

```python
print(data.head())
```

```python
# Combine text columns
data["text"] = data["Make"] + " " + data["Model"] + " " + data["Engine Fuel Type"] + " " + data["Vehicle Style"]

texts = data["text"].dropna().head(100)
```

```python
# Bag of Words
# -------------------
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(texts)

print("Bag of Words")
print(bow.toarray())
```

```python
# Normalized Count
# -------------------
normalized = bow.toarray() / bow.toarray().sum(axis=1, keepdims=True)

print("\nNormalized Count")
print(normalized)
```

```python
# TF-IDF
# -------------------
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(texts)

print("\nTF-IDF")
print(tfidf_matrix.toarray())
```

```python
# Word2Vec
# -------------------
tokenized = [sentence.split() for sentence in texts]

model = Word2Vec(tokenized, vector_size=50, window=3, min_count=1)

print("\nWord2Vec for 'bmw'")
print(model.wv['bmw'])
```
