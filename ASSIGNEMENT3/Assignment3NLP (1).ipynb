```python
# Import libraries
import pandas as pd
import nltk
import re
import pickle

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
data = pd.read_pickle("C:\\Users\\Shree\\Downloads\\News_dataset.pickle")
```

```python
print(data.head())
```

```python
texts = data['Content']
labels = data['Category']
```

```python
# Text Cleaning
# -----------------------------
def clean_text(text):
    text = text.lower()                  # lowercase
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # remove punctuation/numbers
    return text

texts = texts.apply(clean_text)
```

```python
# Remove Stopwords
# -----------------------------
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

texts = texts.apply(remove_stopwords)
```

```python
# Lemmatization
# -----------------------------
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    words = text.split()
    words = [lemmatizer.lemmatize(w) for w in words]
    return " ".join(words)

texts = texts.apply(lemmatize_text)
```

```python
# Label Encoding
# -----------------------------
encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(labels)

# -----------------------------
# TF-IDF Representation
# -----------------------------
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(texts)

print("TF-IDF Shape:", tfidf_matrix.shape)
```

```python
# Save Outputs
# -----------------------------
pd.DataFrame(tfidf_matrix.toarray()).to_csv("tfidf_output.csv", index=False)

pd.DataFrame(encoded_labels, columns=["Encoded_Label"]).to_csv("encoded_labels.csv", index=False)

print("Outputs saved successfully!")
```
