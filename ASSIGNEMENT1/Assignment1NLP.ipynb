```python
from nltk.tokenize import WhitespaceTokenizer, wordpunct_tokenize, TreebankWordTokenizer, TweetTokenizer
from nltk.tokenize import MWETokenizer
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

```python
sentence = "Natural Language Processing is interesting and students are learning NLP easily!"

print("Original Sentence:")
print(sentence)

# 1. Whitespace Tokenization
wt = WhitespaceTokenizer()
print("\nWhitespace Tokenization:")
print(wt.tokenize(sentence))

# 2. Punctuation Based Tokenization
print("\nPunctuation Based Tokenization:")
print(wordpunct_tokenize(sentence))

# 3. Treebank Tokenization
tt = TreebankWordTokenizer()
print("\nTreebank Tokenization:")
print(tt.tokenize(sentence))

# 4. Tweet Tokenization
tweet = TweetTokenizer()
print("\nTweet Tokenization:")
print(tweet.tokenize(sentence))

# 5. MWE Tokenization
mwe = MWETokenizer([('Natural','Language'), ('Language','Processing')])
tokens = sentence.split()
print("\nMWE Tokenization:")
print(mwe.tokenize(tokens))
```

```python
ps = PorterStemmer()
ss = SnowballStemmer("english")

words = sentence.split()

print("\nPorter Stemmer:")
for word in words:
    print(word, "->", ps.stem(word))

print("\nSnowball Stemmer:")
for word in words:
    print(word, "->", ss.stem(word))

# Lemmatization
lemmatizer = WordNetLemmatizer()

print("\nLemmatization:")
for word in words:
    print(word, "->", lemmatizer.lemmatize(word))
```
